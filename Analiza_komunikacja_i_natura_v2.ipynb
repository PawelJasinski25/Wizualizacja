```python
def plot_city_data(df, cities, category_ids, category_name, leg, color1, color2='b'):
    fig, axs = plt.subplots(1, len(cities), figsize=(20, 6), sharey=True, gridspec_kw={'wspace': 0.3})

    for i, city in enumerate(cities):
        handles = []  # Collect handles for legend
        for j, id in enumerate(category_ids):
          df_city = df[(df['cityName'] == city) & (df['categoryId'] == id) & (df['year'] >= 2003) & (df['year'] <= 2022)]

          ax1 = axs[i]
          line, = ax1.plot(df_city['year'], df_city['value'], color=color1[j], marker='o', linestyle='-',  label=category_name)
          handles.append(line)
          ax1.set_title(city)
          ax1.set_xlabel('Rok')
          ax1.set_ylabel(category_name)
          ax1.grid(True)
          ax1.xaxis.set_major_locator(plt.MaxNLocator(integer=True))
          ax1.tick_params(axis='x', rotation=45)

        ax2 = ax1.twinx()
        df_population_change = df[(df['cityName'] == city) & (df['categoryName'] == 'zmiana liczby ludności na 1000 mieszkańców') & (df['year'] >= 2003) & (df['year'] <= 2022)]
        line2, = ax2.plot(df_population_change['year'], df_population_change['value'], color=color2, marker='o', linestyle='-', label='Zmiana liczby ludności')
        handles.insert(0, line2)
        ax2.set_ylabel('Zmiana liczby ludności na 1000 mieszkańców')
        ax2.tick_params(axis='y', labelcolor=color2)

    plt.subplots_adjust(wspace=0.5)
    plt.legend(handles=handles, labels=leg, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
```

```python
def plot_city_data2(df, cities, category_ids, category_name, leg, color1):
    fig, axs = plt.subplots(1, len(cities), figsize=(20, 6), sharey=True, gridspec_kw={'wspace': 0.3})

    for i, city in enumerate(cities):
        handles = []  # Collect handles for legend
        for j, id in enumerate(category_ids):
          df_city = df[(df['cityName'] == city) & (df['categoryId'] == id) & (df['year'] >= 2003) & (df['year'] <= 2022)]

          ax1 = axs[i]
          line, = ax1.plot(df_city['year'], df_city['value'], color=color1[j][0], alpha=1,  label=category_name, zorder=2)
          ax1.fill_between(df_city['year'], df_city['value'], color=color1[j][1],alpha=0.2, zorder=2)
          handles.append(line)
          ax1.set_title(city)
          ax1.set_xlabel('Rok')
          ax1.set_ylabel(category_name)
          ax1.grid(True)
          ax1.xaxis.set_major_locator(plt.MaxNLocator(integer=True))
          ax1.tick_params(axis='x', rotation=45)

        ax2 = ax1.twinx()
        df_population_change = df[(df['cityName'] == city) & (df['categoryName'] == 'ludność w tysiącach') & (df['year'] >= 2003) & (df['year'] <= 2022)]
        line2, = ax2.plot(df_population_change['year'], df_population_change['value'], color='b', alpha=1, label='Liczba mieszkańców (tys.)')
        ax2.fill_between(df_population_change['year'], df_population_change['value'], color='dodgerblue',alpha=0.2)
        handles.insert(0, line2)
        ax2.set_ylabel('Liczba mieszkańców (tys.)')
        ax2.tick_params(axis='y', labelcolor='b')

    plt.subplots_adjust(wspace=0.5)
    plt.legend(handles=handles, labels=leg, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
```

```python
cities = ['Marki', 'Ząbki', 'Piaseczno', 'Bytom', 'Sopot', 'Sosnowiec']
ludnosc = 458603
other_categories = [73854, 73855, 2037, 288080]
cat_names = ["parki spacerowo - wypoczynkowe", "zieleńce", "Powierzchnia gruntów leśnych", "Długość dróg rowerowych"]
#other_categories = [73854, 73855, 2037, 1365258, 288080]
#cat_names = ["parki spacerowo - wypoczynkowe", "zieleńce", "Powierzchnia gruntów leśnych", "Czynne przystanki komunikacji miejskiej", "Długość dróg rowerowych"]

# Filter for the specific cities
filtered_df = df[df['cityName'].isin(cities)]

pivot_df = filtered_df.pivot_table(index=['cityName', 'year'], columns='categoryId', values='value').reset_index()
```

```python
correlations = {}

for city in cities:
    city_data = pivot_df[pivot_df['cityName'] == city]

    city_correlations = {}
    for category in other_categories:
        # Get overlapping years where both category_a and the other category have data
        valid_years = city_data.dropna(subset=[ludnosc, category])

        if not valid_years.empty:
            correlation = valid_years[[ludnosc, category]].corr().iloc[0, 1]
            city_correlations[category] = correlation
        else:
            city_correlations[category] = float('nan')  # or use another placeholder for missing correlations

    correlations[city] = city_correlations

correlations
```

```python
plot_data = pd.DataFrame(correlations).T
plot_data = plot_data.fillna(0.01)

# Plotting
plot_data.plot(kind='bar', figsize=(10, 6))
plt.title('Korelacje pomiędzy kategoriami a zmianą liczby ludności')
plt.xlabel('Miasta')
plt.ylabel('Współczynnik korelacji')
plt.legend(labels=cat_names,title='Kategorie', loc='lower right')
plt.axhline(y=0, color='black', linewidth=1)
plt.show()
```
