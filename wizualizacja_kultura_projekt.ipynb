```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/PawelJasinski25/Wizualizacja/main/over_30k.csv'

response = requests.get(url)
with open('over_30k.csv', 'wb') as f:
    f.write(response.content)

df = pd.read_csv('over_30k.csv')
print(df)
```

```python
url2 = 'https://raw.githubusercontent.com/PawelJasinski25/Wizualizacja/main/average_change.csv'

response = requests.get(url2)
with open('average_change.csv', 'wb') as f:
    f.write(response.content)

df2 = pd.read_csv('average_change.csv')
print(df2)
```

```python
najmniejszy=df2.tail(8)
najwiekszy =df2.head(7)
#najset=pd.concat([najwiekszy,najmniejszy])
najmniejszy=najmniejszy.iloc[-8:-1]
print(najmniejszy)
print(najwiekszy)
```

```python
najmniejszy = najmniejszy['cityName'].tolist()
print(najmniejszy)
najwiekszy = najwiekszy['cityName'].tolist()
print(najwiekszy)
kultura = ["Biblioteki publiczne", "Muzea", "Kina", "Instytucje muzyczne", "Teatry", "Imprezy artystyczne","Domy kultury"]
filtrowane_najwieksze=df[df['cityName'].isin(najwiekszy)]
filtrowane_najmniejsze=df[df['cityName'].isin(najmniejszy)]
```

```python
biblioteki_set_mn=filtrowane_najmniejsze[filtrowane_najmniejsze['categoryName']==kultura[0]]
biblioteki_set_w=filtrowane_najwieksze[filtrowane_najwieksze['categoryName']==kultura[0]]
muzea_set_mn=filtrowane_najmniejsze[filtrowane_najmniejsze['categoryName']==kultura[1]]
kina_set_mn=filtrowane_najmniejsze[filtrowane_najmniejsze['categoryName']==kultura[2]]
instytucje_set_mn=filtrowane_najmniejsze[filtrowane_najmniejsze['categoryName']==kultura[3]]
teatry_set_mn=filtrowane_najmniejsze[filtrowane_najmniejsze['categoryName']==kultura[4]]
imprezy_set_mn=filtrowane_najmniejsze[filtrowane_najmniejsze['categoryName']==kultura[5]]
muzea_set_w=filtrowane_najwieksze[filtrowane_najwieksze['categoryName']==kultura[1]]
kina_set_w=filtrowane_najwieksze[filtrowane_najwieksze['categoryName']==kultura[2]]
instytucje_set_w=filtrowane_najwieksze[filtrowane_najwieksze['categoryName']==kultura[3]]
teatry_set_w=filtrowane_najwieksze[filtrowane_najwieksze['categoryName']==kultura[4]]
imprezy_set_w=filtrowane_najwieksze[filtrowane_najwieksze['categoryName']==kultura[5]]
domk_set_mn=filtrowane_najmniejsze[filtrowane_najmniejsze['categoryName']==kultura[6]]
domk_set_w=filtrowane_najwieksze[filtrowane_najwieksze['categoryName']==kultura[6]]
```

## **Czy ilość bibliotek publicznych ma wpływ?**



```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=biblioteki_set_mn, x='year', y='value', hue='cityName', marker='o')

plt.title('Ilość bibliotek publicznych w miastach, gdzie ludność spada najwięcej')
plt.xlabel('Rok')
plt.ylabel('Wartość')
plt.legend(title='Miasto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
kategorie = ["zmiana liczby ludności na 1000 mieszkańców","Biblioteki publiczne"]
porownanie = df[df['categoryName'].isin(kategorie) & (df['cityName'] == 'Łódź')]

plt.figure(figsize=(12, 8))

kolory = {
    "Biblioteki publiczne": "#9FC490",
    "zmiana liczby ludności na 1000 mieszkańców":"#465362"
}

for k in porownanie['categoryName'].unique():
    data = porownanie[porownanie['categoryName'] == k]
    plt.bar(data['year'],data['value'], label=k,color=kolory[k])

plt.xlabel('Rok')
plt.ylabel('Ilość')
plt.title('Porównanie liczby bibliotek publicznych, a zmiany ludności na 1000 osób dla Łodzi')
plt.legend(title='Kategoria')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=biblioteki_set_w, x='year', y='value', hue='cityName', marker='o')

plt.title('Ilość bibliotek publicznych w miastach, gdzie ludność wzrasta najwięcej')
plt.xlabel('Rok')
plt.ylabel('Wartość')
plt.legend(title='Miasto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
porownanie = df[df['categoryName'].isin(kategorie) & (df['cityName'] == 'Zielona Góra')]

plt.figure(figsize=(12, 8))

kolory = {
    "zmiana liczby ludności na 1000 mieszkańców":"#465362",
    "Biblioteki publiczne": "#9FC490"
}

for k in porownanie['categoryName'].unique():
    data = porownanie[porownanie['categoryName'] == k]
    plt.bar(data['year'],data['value'], label=k,color=kolory[k])

plt.xlabel('Rok')
plt.ylabel('Ilość')
plt.title('Porównanie liczby bibliotek publicznych, a zmiany ludności na 1000 osób dla Zielonej Góry')
plt.legend(title='Kategoria')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

Nie ma, bo w miastach, w których ludność spadła najbardziej jest dużo więcej bibliotek niż w tych, których znacznie wzrosła
